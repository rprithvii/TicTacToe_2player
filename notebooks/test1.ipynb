```python
class Seat:
    def __init__(self, seat_row, seat_col):
        self.seat_row = seat_row
        self.seat_col = seat_col
        self.is_booked = False

    def __str__(self):
        if self.is_booked == True:
            return "[X]"
        else:
            return "[ ]"
```

```python
class Cinema:
    def __init__(self, n_rows_cinema, n_cols_cinema):
        self.n_rows_cinema = n_rows_cinema
        self.n_cols_cinema = n_cols_cinema

        self.seats = []
        for i in range(1, n_rows_cinema + 1):
            for j in range(1, n_cols_cinema + 1):
                self.seats.append(Seat(i, j))

    def display_cinema(self):
        for index, j in enumerate(self.seats):
            if (index+1)%self.n_cols_cinema == 0:
                print(j, end = "\n")
            else:
                print(j, end = " ")        
     
    def book_ticket(self, chosen_row, chosen_col):
        for i in self.seats:
            if i.seat_row == chosen_row and i.seat_col == chosen_col and i.is_booked == True:
                print(f"Chose another seat. It is already booked")
            elif i.seat_row == chosen_row and i.seat_col == chosen_col:
                i.is_booked = True
                print(f"You chosen seat is booked. Seat Number: R{chosen_row}C{chosen_col}")    
```

```python
c = Cinema(3, 3)
```

```python
c.display_cinema()
```

```python
c.book_ticket(2,2)
```

```python
class Driver:
    def __init__(self, driver_name, car_model, is_available = True):
        self.driver_name = driver_name
        self.car_model = car_model
        self.is_available = is_available
```

```python
class Passenger:
    def __init__(self, passenger_name, wallet_balance):
        self.passenger_name = passenger_name
        self.wallet_balance = wallet_balance
```

```python
class Ride:
    ride_id_counter = 10000
    def __init__(self, passenger_obj, driver_obj, ride_distance):
        self.ride_id = Ride.ride_id_counter
        Ride.ride_id_counter = Ride.ride_id_counter + 1
        self.passenger_obj = passenger_obj
        self.driver_obj = driver_obj
        self.ride_distance = ride_distance
        self.minimum_cost = 4.0
        self.rate_per_km = 2.1

    def estimate_ride_fare(self):
        estimated_fare = self.minimum_cost + self.ride_distance*self.rate_per_km
        return estimated_fare

    def start_ride(self):
        if self.estimate_ride_fare() <= self.passenger_obj.wallet_balance and self.driver_obj.is_available:
            print(f"Your ride has been booked! \n"
                  f"Ride id: {self.ride_id}\n"
                  f"{self.driver_obj.driver_name} is on the way with a {self.driver_obj.car_model} \n"
                  f"Your estimated fare is: ${round(self.estimate_ride_fare(),2)}")  
        elif self.driver_obj.is_available == False:
            print(f"Your booking can't be completed as driver is not available")

        elif self.estimate_ride_fare() > self.passenger_obj.wallet_balance:
            print(f"Your booking can't be completed as you have insufficient balance in wallet")

    def end_ride(self):
        actual_fare = np.random.random() + self.estimate_ride_fare()*1.05
        print(f"Ride is completed.\n"
        f"Estimated fare: ${round(self.estimate_ride_fare(),2)}. \n"
        f"Actual fare: £{round(actual_fare,2)}")
```

```python
class Component:
    def __init__(self, part_name, category, power_draw):
        self.part_name = part_name
        self.category = category
        self.power_draw = power_draw
```

```python
class Computer:
    def __init__(self, model_name):
        self.model_name = model_name
        self.psu_watts = 0 
        self.parts = []

    def install_part(self, component_obj):
        self.parts.append(component_obj)
        print(f"{component_obj.part_name} has been installed !"
        # f"Details:\n"
        # f"--------\n"
        # f"Part name: {component_obj.part_name}\n"
        # f"Category: {component_obj.category}\n"
        # f"Power: {component_obj.power_draw}\n"
        )

    def get_installed_parts(self):
        print(f"Instaled parts are:")
        for i in self.parts:
            print(f"{i.part_name}, {i.category}, {i.power_draw}")
            # self.psu_watts = self.psu_watts + i.power_draw
        # print(f"Total power: {self.psu_watts}")
        

    def get_computer_power(self):
        for i in self.parts:
            # print(self.psu_watts, i.power_draw)
            self.psu_watts = self.psu_watts + i.power_draw
        return self.psu_watts
```

```python
class Member:
    class_id_counter = 10000
    def __init__(self, name, balance):
        
        self.member_id = Member.class_id_counter + 1
        self.name = name
        self.balance = balance

        Member.class_id_counter = Member.class_id_counter + 1
```

```python
class Gym:
    def __init__(self):
        self.member_list = []

    def add_member(self, member_instance):
        self.member_list.append(member_instance)

    def show_all_members(self):
        for i in self.member_list:
            print(f"Member_id:, {i.member_id}, Name: {i.name}, Balance: {i.balance}")

    def add_multiple_members(self, list1):
        for i in list1:
            self.add_member(i)

    def show_member_details_by_name(self, member_name):
        for i in self.member_list:
            if i.name == member_name:
                print(f"Member_id:, {i.member_id}, Name: {i.name}, Balance: {i.balance}")
```

```python
m1 = Member('Rohan', 4000)
m2 = Member('Sohan', 300)
m3 = Member('Rohit', 5000)
```

```python
g = Gym()
```

```python
g.add_multiple_members(list1 = [m1, m2, m3])
```

```python
g.show_all_members()
```

```python
g.show_member_details_by_name('Rohan')
```

```python
class Book:
    def __init__(self, name, author):
        self.name = name
        self.author = author
        self.is_available = True
```

```python
class Library:
    def __init__(self):
        self.book_list = []

    def add_book(self, book_instance):
        self.book_list.append(book_instance)
        print(f"'{book_instance.name}' from '{book_instance.author}' is added to Library")

    def get_book_inventory(self):
        print(f"These are the books present in Library:")
        for i in range(len(self.book_list)):
            print(f"{i+1}. {self.book_list[i].name} by {self.book_list[i].author}")

    def lend_book(self, book_name_to_be_lended):
        book_exists_check = 0
        for i in self.book_list:
            if i.name == book_name_to_be_lended:
                print(f"Your book exists in Library. Here is your copy")
                self.book_list.remove(i)
                book_exists_check = book_exists_check + 1
        if book_exists_check == 0:
            print(f"The requested book does not exist in the library")
```

```python
b1 = Book('Harry Potter', 'J K Rowling')
b2 = Book('Harry Potter 1', 'J K Rowling')
b3 = Book('Mystery', 'Dan Brown')
```
